```python
import scipy as sp
import math as m
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import figure
from scipy.optimize import curve_fit
import pylab as py

gradiente_velocidad = np.array([50, 70, 90, 110, 130]) #Arreglo de datos de los valores del gradiente de velocidad. El argumento de np.array es una lista []
esfuerzo_cortante = np.array([6.01, 7.48, 8.59, 9.19, 10.21])

def funj(g,m,n):
    return m*g**n

parametros, pcov = curve_fit(funj, gradiente_velocidad, esfuerzo_cortante)

t = np.linspace(gradiente_velocidad[0], gradiente_velocidad[-1])

plt.plot(gradiente_velocidad, esfuerzo_cortante, "ok", label='datos')
plt.plot(t,funj(t,*parametros), 'r--')
plt.xlabel("Gradiente de velocidad $\gamma$ [1/s]")
plt.ylabel("Esfuerzo cortante $\\tau$ [N/$m^2$]")
plt.grid()

```
