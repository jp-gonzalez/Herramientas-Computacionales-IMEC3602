## 2. Implementación de una función que permita graficar la distribución de energía y comparación entre diferentes series

Luego de haber importado los datos, se procede a implementar una función que permita graficar la distribución de energía y comparación entre diferentes series. Lo primero es definir las funciones ghi_mensual y ghi_dia, las cuales retornan cada una un DataFrame (ghi_df) con los valores de la radiación promedio mensual y diaria, respectivamente.

```python
def ghi_mensual(serie_horaria):
    agnos = sorted(serie_horaria['AÑO'].unique())
    ghi_df = pd.DataFrame(columns=["AÑO", "MES", "DIAS", "GHImes"])
    for agno in agnos:
        for mes in range(1,13):
            nombre_energia = list(serie_horaria)[-1]
            dias = sorted(serie_horaria[(serie_horaria['AÑO'] == agno) & (serie_horaria['MES'] == mes)]["DIA"].unique())
            ghi_mes = serie_horaria.loc[((serie_horaria["AÑO"] == agno) & (serie_horaria["MES"] == mes)), nombre_energia].sum()/1000
            ghi_dict = {"AÑO":agno, "MES":mes, "DIAS":dias[-1],
                        "GHImes":ghi_mes}
            ghi_df = ghi_df.append(ghi_dict, ignore_index=True)
    return ghi_df
```

```python
def ghi_dia(serie_ghi_mensual):
    agnos = sorted(serie_ghi_mensual['AÑO'].unique())
    ghi_df = pd.DataFrame(columns=["AÑO", "MES", "GHIdiario"])
    for agno in agnos:
        for mes in range(1,13):
            dias = serie_ghi_mensual.loc[((serie_ghi_mensual["AÑO"] == agno)
                            & (serie_ghi_mensual["MES"] == mes))
                                         ,"DIAS"].item()
            ghi_diario = serie_ghi_mensual.loc[(
                (serie_ghi_mensual["AÑO"] == agno)
                            & (serie_ghi_mensual["MES"] == mes)),
                "GHImes"].item()/dias
            ghi_dict = {"AÑO":agno, "MES":mes, "GHIdiario":ghi_diario}
            ghi_df = ghi_df.append(ghi_dict, ignore_index=True)
    return ghi_df
```

Aplicando las funciones anteriores a los DataFrame correspondientes (serie_pd_tierra y serie_pd_satelital), se obtienen dos listas con los valores de la radiación ordenados de menor a mayor aplicando la función sorted(). De esta manera, la lista datos_IDEAM representa la lista del promedio mensual diario de la radiación de los datos en tierra, y la lista datos_NREL representa la lista del promedio mensual diario de la radiación de los datos del satélite.

```python
# Aplicación de las funciones para los datos en la tierra
ghi_mensual_loc_tierra = ghi_mensual(serie_pd_tierra)
ghi_diario_loc_tierra = ghi_dia(ghi_mensual_loc_tierra)
datos_IDEAM = sorted(ghi_diario_loc_tierra["GHIdiario"]);

# Aplicación de las funciones para los datos satelitales
ghi_mensual_loc_satelite = ghi_mensual(serie_pd_satelite)
ghi_diario_loc_satelite = ghi_dia(ghi_mensual_loc_satelite)
datos_NREL = sorted(ghi_diario_loc_satelite["GHIdiario"])
```

## 3. Implementación de una función para el cálculo de la ENFICC

Luego esto, se procede a calcular el valor de la ENFICC(99%) con base en los datos obtenidos para las dos series de datos (IDEAM y NREL). Como se puede esperar, los valores de la ENFICC van a ser diferentes para las dos series, por lo cual se procederá a calcular la diferenciar porcentual de dichas valores para los análisis. Dicha diferencia porcentual se piensa reducir con el método del ajuste MCP, de tal forma que se calculará un nuevo valor de la diferencia de la ENFICC después de aplicar dicho ajuste en el iteral 4. Así, la función de la enficc es la siguiente:

```python
def enficc(s_ideam, s_nrel):
    percentil = 0.01
    enficc_tierra = s_ideam.quantile(q=percentil, 
                                 numeric_only=True)["GHIdiario"]
    enficc_satelite = s_nrel.quantile(q=percentil, 
                                 numeric_only=True)["GHIdiario"]   
    print('La ENFICC 99% de la serie IDEAM es: ', "%.2f" % enficc_tierra)
    print('La ENFICC 99% de la serie NREL es: ', "%.2f" % enficc_satelite)
    dp = ((enficc_satelite- enficc_tierra)/enficc_tierra)*100
    print('La diferencia porcentual de las ENFICC es [%]: ', "%.2f" % dp)
```

Luego, se procede a aplicar dicha función. En este caso, se espera que los dos valores de la ENFICC sean muy diferentes, ya que, como se explico anteriormente, los datos no han sido ajustados por el método MCP. Dicho esto, los resultados para la ENFICC en las dos series son los siguientes:

```python
enficc(ghi_diario_loc_tierra, ghi_diario_loc_satelite)
```

Calculando la diferencia porcentual entre estos (la cual se define en este caso como: $DP[\%] = (enficc_i - enficc_n)/(enficc_n)*100 $) se puede apreciar que la serie NREL sobreestima los datos de la radiación del IDEAM en un 30.92$\%$. Este valor se analizará nuevamente luego del ajuste MCP.

## 4. Implementación de una función para la metodología MCP basada en relación de varianzas

Como bien se sabe, la metodología MCP se aplica a partir de un número de meses determinado. Esto significa que es necesario seccionar la serie para que se obtenga un DataFrame desde un mes inicial hasta un mes final. De esta forma, se define la función serie_corta, la cual tiene como entradas el DataFrame a seccionar, el año en el cual se eligen los meses para el ajuste, el mes inicial y el mes final. Luego de aplicar la función, esta retorna un nuevo DataFrame (ghi_df) con la serie corta.

```python
def serie_corta(serie_h_tierra, agno, mes_inicio, mes_final):
    corta=[]
    for mes in range(mes_inicio,mes_final+1):
        datos=serie_h_tierra.loc[(serie_h_tierra["AÑO"] == agno)& (serie_h_tierra["MES"] == mes)]
        corta.append(datos)
        ghi_df = pd.concat(corta)
            
    return ghi_df
```

Debido a que se desea realizar varios ajustes con diferentes rangos de meses en algunos años, esta función se va a aplicar más de una vez. En total se aplicó la función 6 veces, en el cual se realizó el ajuste del error en los años 2009 y 2013 para los siguientes rangos de meses: 1 a 3 meses, 1 a 6 meses, y 1 a 12 meses.

```python
# AJUSTES PARA EL AÑO 2009
serie_corta_1 = serie_corta(serie_pd_tierra, 2009, 1, 3)
serie_corta_2 = serie_corta(serie_pd_tierra, 2009, 1, 6)
serie_corta_3 = serie_corta(serie_pd_tierra, 2009, 1, 12)
# AJUSTES PARA EL AÑO 2013
serie_corta_4 = serie_corta(serie_pd_tierra, 2013, 1, 3)
serie_corta_5 = serie_corta(serie_pd_tierra, 2013, 1, 6)
serie_corta_6 = serie_corta(serie_pd_tierra, 2013, 1, 12)
```

Con esta función definida, se procede a hacer la función del método de ajuste MCP, cuyas entradas son: la serie seccionada en formato horario, y la serie satelital en este mismo formato. En este caso, la salida es un DataFrame de nombre no_negativos que devuelve la serie satelital ajustada. Esta función también elimina los valores negativos de la radiación ya que el ajuste MCP genera valores menores a cero para aquellos datos que no registran radiación alguna (datos tomados en la noche).

```python
def mcp(serie_tierra_c, serie_satelital):
    
    satelital_corta = []
    agno = int(serie_tierra_c['AÑO'][0:1])
    mes_inicio = int(serie_tierra_c['MES'][0:1])
    mes_final = int(serie_tierra_c['MES'][-2:-1]) 
    ghi_df = serie_corta(serie_satelital, agno, mes_inicio, mes_final)
    satelital_modificada = serie_satelital.copy(deep=True)
    
    media_tierra = serie_tierra_c['IDEAM'].mean()
    desv_tierra = np.std(serie_tierra_c['IDEAM'])

    media_satelite = ghi_df['NREL'].mean()
    desv_satelite = np.std(ghi_df['NREL'])
    
    m = (desv_tierra/desv_satelite)
    satelital_modificada['NREL'] = (media_tierra - m*media_satelite) + m*(serie_satelital['NREL'])
    
    offset = float(satelital_modificada['NREL'][0:1]) 
    no_negativos = satelital_modificada.replace(offset,0)
           
    return no_negativos
```

Con la función realizada, se procede a aplicar el ajuste mcp a los datos satelites con las 6 series de datos mencionadas anteriormente. De esta manera, se definen los DataFrame con nombre mcp, los cuales tienen la radiación satelital ajustada de los 10 años en formato horario.

```python
# AJUSTES PARA EL AÑO 2009
mcp_1 = mcp(serie_corta_1, serie_pd_satelite);mcp_2 = mcp(serie_corta_2, serie_pd_satelite);mcp_3 = mcp(serie_corta_3, serie_pd_satelite)
# AJUSTES PARA EL AÑO 2013
mcp_4 = mcp(serie_corta_4, serie_pd_satelite);mcp_5 = mcp(serie_corta_5, serie_pd_satelite);mcp_6 = mcp(serie_corta_6, serie_pd_satelite)
```

## 5. Implementación de una función para el cálculo de de los errores planteados: RMS, MBE y KSI (P-Value)

Finalmente, el último paso consiste en calcular los errores  RMS, MBE y KSI (PValue). Con estos errores es posible evaluar la precisión del ajuste realizado en el iteral anterior. Así, se definió la función -errores-, la cual tiene como entrada 2 dataFrame (datos de la IDEAM y datos NREL corregidos) y como salida 1 lista que devuelve los errores calculados. En dicha lista, los errores se ordenan de la siguiente manera: errores[0] = RMS, errores[1] = MBE y errores[2] = KSI (PValue).

```python
def errores(ideam,corregida):
    corregida_corta = []
    agno = int(ideam['AÑO'][0:1])
    mes_inicio = int(ideam['MES'][0:1])
    mes_final = int(ideam['MES'][-2:-1]) 
    for mes in range(mes_inicio,(mes_final+1)):
        datos_c = corregida.loc[(corregida["AÑO"] == agno)& (corregida["MES"] == mes)]
        corregida_corta.append(datos_c)
        dfc = pd.concat(corregida_corta) 
    a = list(ideam['IDEAM'])
    b = list(dfc['NREL'])
    datos_id = pd.DataFrame(a)
    datos_nrel = pd.DataFrame(b)
    rmsen = ((m.sqrt(((datos_id[0]-datos_nrel[0])**2).mean()))/(max(datos_id[0])-min(datos_nrel[0])))*100
    MBE = abs((((datos_nrel[0] - datos_id[0]).mean())*len(datos_id[0])/(datos_id[0].mean()*len(datos_id[0]))))*100
    pi_value = stats.ks_2samp(datos_id[0],datos_nrel[0])
    errores = [rmsen, MBE, pi_value[1]*100]
    return errores
```

```python
# EVOLUCIÓN DE ERRORRES PARA EL AÑO 2009
eh_1 = errores(serie_corta_1, mcp_1);eh_2 = errores(serie_corta_2, mcp_2);eh_3 = errores(serie_corta_3, mcp_3)
# EVOLUCIÓN DE ERRORRES PARA EL AÑO 2013
eh_4 = errores(serie_corta_4, mcp_4);eh_5 = errores(serie_corta_5, mcp_5);eh_6 = errores(serie_corta_6, mcp_6)
```
